#### le code de recuperation des données

```python
import requests
url = "https://jsonplaceholder.typicode.com/users"
response = requests.get(url)
print(response.status_code)
```

```python
users = response.json()
for user in users:
    print(user["id"], user["name"], user["email"])
    
```

#### Extration des données selon le besoin

```python
# tranformation des donnée
data = []     # defini une base de donnée vide
print(user)
#print(data)
```

```python
for u in users:
    data.append({
        "id":u["id"],
        "name":u["name"],
        "email":u["email"],
        "city":u["address"]["city"],
    })
print(data)
```

#### conversion des données en DATAFRAME

```python
import pandas as pd
base = pd.DataFrame(data)
print(base)
```
